```python
# Chlorophyll

import xarray as xr

myts = pd.Series(pd.date_range("1998-01-01", "2019-12-31", freq="MS"))
#myts = pd.Series(pd.date_range("1998-01-01", "1999-01-01", freq="MS"))
for i in myts:
    
    # Returns dates in a given month
    days   = pd.Series(pd.date_range(i, periods=i.days_in_month, freq="D"))
    
    # Convert to days since 1-1-1970
    days = ((days - pd.datetime(1970,1,1)) / np.timedelta64(1, 'D')).astype(int)
    
    # Find these days in time array
    tind  = np.argwhere( (dtime>=min(days)) & (dtime<=max(days))).ravel()
  
    url = u'https://rsg.pml.ac.uk/thredds/dodsC/CCI_ALL-v4.2-DAILY?lat[1164:1319],lon[2472:2639],time['+ str(min(tind)) + ':' + str(max(tind)) + '], \
chlor_a[' + str(min(tind)) + ':' + str(max(tind)) + '][1164:1319][2472:2639]'


    print(url)
    
    ds = xr.open_dataset(url, engine='netcdf4')
    ds.to_netcdf('/media/silsbelab/SILSBE/mab/cci/chl/chl_' + i.strftime('%Y%m') + '.nc')
```
